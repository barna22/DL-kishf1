```python
# I only managed to do this part of the task for 1 image instead of 5

import cv2
import numpy as np
import matplotlib.pyplot as plt

#read image
img = cv2.imread('hmm.jpg', cv2.IMREAD_COLOR)

#calculate histograms for the 3 channels, with full mask
for i in range(3):
    plt.plot(cv2.calcHist([img], [i], None, [256], [0, 256]))

#the histogram isn't correct, unfortunately couldn't figure out why
plt.rcParams['figure.figsize'] = [5, 5]
plt.show()

#calculate mean and std before normalization
mean, std = cv2.meanStdDev(img)
print('Means, standard deviations before normalization:\n', mean, std)

newmean, newstd = [], []
#get the three channels from the image
channels = cv2.split(img)

#calculate new mean and std values for the normalized channel
for i in range(3):
    norm = (channels[i] - mean[i]) / std[i]
    newmean.append(float(cv2.meanStdDev(norm)[0]))
    newstd.append(float(cv2.meanStdDev(norm)[1]))

print('Means, standard deviations after normalization:\n', newmean, newstd)
```

```python
import matplotlib.pyplot as plt
from scipy import signal
from scipy.io import wavfile

#similar to matlab audioread :)
#FS is sampling frequency
FS, data1 = wavfile.read('chime.wav')
#create spectrogram plot with FFT size
plt.specgram(data1, Fs=FS, NFFT=128, noverlap=0)
plt.show()

FS, data2 = wavfile.read('guitar.wav')
plt.specgram(data2, Fs=FS, NFFT=128, noverlap=0)
plt.show()
```

```python
#calculate mean and std of first sample
mean = np.mean(data1)
std = np.std(data1)
#normalization step, won't be perfect due to numerical error
data1_normalized = (data1 - mean) / std
print('Mean of first sample: ', np.mean(data1_normalized))
print('STD of first sample : ', np.std(data1_normalized))

#calculate mean and std of second sample
mean = np.mean(data2)
std = np.std(data2)
data2_normalized = (data2 - mean) / std
print('Mean of second sample: ', np.mean(data2_normalized))
print('STD of second sample : ', np.std(data2_normalized))
```

```python
from collections import Counter

#load the text file
f = open('oktatas-deep-learning')
text = f.readlines()

#write the contents of the list into a single string and print it
full_text = ' '.join(text)
print(full_text)

#get the number of occurences for each character
occurrences = Counter(full_text)
#remove whitespaces and new lines from the character dictionary
occurrences.pop(' ')
occurrences.pop('\n')

#create plot and increase its size so that characters are visible
plt.bar(list(occurrences.keys()), list(occurrences.values()), color='g')
plt.rcParams['figure.figsize'] = [20, 30]
plt.show()
```
